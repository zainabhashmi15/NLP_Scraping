```python
from bs4 import BeautifulSoup
import requests
```

```python
urls = ["https://www.amazon.in/Airdopes-121v2-Bluetooth-Immersive-Assistant/product-reviews/B08JQN8DGZ/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"]
```

```python
#iterate through the urls
for url in urls:
  #send a GET request to the url
  response = requests.get(url)

  #check if  the request was succesfull
  if response.status_code == 200:
    #parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content,'html.parser')
    print(soup)
```

<h1><b> Data Pre-processing </b></h1>

<li> import regex as re </li>
<li> This is explained below [^\w\s] </li>
<li> " [ " character block start </li>
<li> " ^ " not these characters (letters, numbers) </li>
<li> " \w " words characters </li>
<li> " \s " space characters </li>
<li> " ] " character block end </li>

```python
import re

modified_review_title = []

for title in df['title']:
    modified_title = re.sub(r'[^\w\s]', '', title)
    modified_title = modified_title.replace('\n', ' ')
    modified_review_title.append(modified_title)
```

<h1><b> Tokenization  </b></h1>
<h6><b> ---> Breaking down of sentenses into words. </b></h6>



---



```python
# Import packages

import nltk
nltk.download('punkt')
nltk.download('wordnet')
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
text = modified_review_title
text
```

```python
# word tokenization
# If we use append it will make list of ith word

word_tokens = []

for words in text:
    tokens = word_tokenize(words)
    word_tokens.extend(tokens)

print('Word Tokens:', word_tokens)
```

```python
# sentence tokenization

sent_tokens = []

for sentence in text:
  tokens = sent_tokenize(sentence)
  sent_tokens.extend(tokens)

print(f'sent _tokens : {sent_tokens}')
```

<h1> <b> POS tagging for words </b> </h1>

```python
import nltk
nltk.download('averaged_perceptron_tagger')

#perform pos tagging
pos_tags = nltk.pos_tag(word_tokens)
print('pos_tags',' : ',pos_tags)

#print the pos tags
for words, tag in pos_tags:
  print('word : ',words,'---->', tag)
```

<h1><b> Word Count </b> </h1>

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

# Tokenize the corpus into individual words
#words = [word for sentence in corpus for word in word_tokenize(sentence)]
words = word_tokenize(' '.join(sent_tokens))

#calculate the word freq distribution
freq_dist = FreqDist(words)

#print the word count for each word
print('word count :')
for word,count in freq_dist.items():
  print(f"{word}: {count}")
```
